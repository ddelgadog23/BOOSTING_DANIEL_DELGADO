```python
# Se procede a eliminar las variables blood preassure y skinthickness

df = df.drop('SkinThickness', axis =1)
df = df.drop('BloodPressure', axis =1)
```

```python
# ANALISIS DE OUTLIERS

df.describe()
```

```python
df.info()
```

```python
df = df[df['Glucose'] > 0 ]
df = df[df['BMI'] > 0 ]
```

```python
df.info()
```

```python
# Analizamos numero de embarazos segun outliers

df[df['Pregnancies'] > 12]
```

```python
df = df[df['Pregnancies']<12]
```

```python
from sklearn.preprocessing import MinMaxScaler
x = df.drop('Outcome',axis =1)

scaler = MinMaxScaler()
df_escalado = pd.DataFrame(scaler.fit_transform(x), index = df.index, columns = x.columns)
df_escalado['Outcome'] = df['Outcome']
```

```python
# SELECCION DE CARACTERISTICAS

from sklearn.model_selection import train_test_split

x = df_escalado.drop('Outcome',axis=1)
y = df_escalado['Outcome']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# CORREMOS EL MODELO


```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

model = XGBClassifier(random_state = 42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from  sklearn.model_selection import GridSearchCV

hyper = {
    'n_estimators': [10,20,30,60,100,200],
    'learning_rate': [100,80,40,20,30],
    'max_depth': [3,10,20,50,100],
    
}

grid = GridSearchCV(model, hyper, scoring = "accuracy", cv = 5)
grid
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(x_train,y_train)

print(f'Los mejores hiperparametros: {grid.best_params_}')
```

```python
model_mejorado = XGBClassifier(learning_rate = 100, max_depth = 10, n_estimators = 10, random_state = 42)
model_mejorado.fit(x_train,y_train)
```

```python
y_pred2 = model_mejorado.predict(x_test)
accuracy_score(y_test,y_pred2)
```

# Analisis

- la precision del modelo bajo, pareciera que esta sobre ajustado.

```python
model_mejorado2 = XGBClassifier(learning_rate = 1, max_depth = 100, n_estimators = 100, random_state = 42)
model_mejorado2.fit(x_train,y_train)

y_pred3 = model_mejorado2.predict(x_test)
accuracy_score(y_test,y_pred3)
```

Al disminuir el learning rate y aumentando el numero de arboles y su profundidad, mejoro la precision.
